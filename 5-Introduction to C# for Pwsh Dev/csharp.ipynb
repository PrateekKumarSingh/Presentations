## Example - Call inline PowerShell

This example demonstrates that you can run PowerShell from C#

```
using System;
using System.Management.Automation;

PowerShell ps = PowerShell.Create();
var output = ps.AddCommand("Get-Process").Invoke();
foreach (PSObject result in output)
{
    if (!String.IsNullOrEmpty(result.Members["ProcessName"].Value.ToString()))
    {
        Console.WriteLine("{0,-50}{1}",
            result.Members["ProcessName"].Value,
            result.Members["Id"].Value);
    }
}
 
```
